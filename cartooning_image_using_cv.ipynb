## IMPORTING LIBRARIES

```python
import cv2
import matplotlib.pyplot as plt
```

# LOADING IMAGE

```python
img=cv2.imread('/Users/saiteja/Desktop/spidy.jpeg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

## CONVERTING IMAGE TO GREY SCALE

```python
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(grey, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

## REDUCING NOISE

```python
grey = cv2.medianBlur(grey,5)
plt.imshow(cv2.cvtColor(grey, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

## CONVERTING INTO BINARY IMAGE

```python
thresh = cv2.adaptiveThreshold(grey, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

```

## SMOOTHING IMAGE

```python
bilateral = cv2.bilateralFilter(img, 19, 75, 75)
plt.imshow(cv2.cvtColor(bilateral, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

```

## CREATING CARTOON 

```python
cartoon = cv2.bitwise_and(bilateral, bilateral, mask=thresh)
plt.imshow(cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
